```python
# Import required packages 

import pandas as pd
import numpy as np
from sklearn import linear_model
import requests
from nba_api.stats import endpoints
from matplotlib import pyplot as plt
```

```python
# Check League Leaders (season totals) module through endpoints, assign said class to "data"

data = endpoints.leagueleaders.LeagueLeaders() 

# The newly created data_frame now has the built-in functions to explore data

df = data.league_leaders.get_data_frame() 
```

```python
# Call head method to check if data has loaded correctly

df.head()
```

```python
df.sort_values("FGA", ascending = False)


# We want to look at the field-goal attemps (FGA) and its relationship to points (PTS). We are expecting a positive correlation (R2).
# Get the stats per game by dividing each variable by games played (GP). This will give us the per-game average

x,y = df.FGA/df.GP, df.PTS/df.GP

# Since we are using a simple linear regression, we need to reshape the np array from 1d to 2d

x = np.array(x).reshape(-1,1)
y = np.array(y).reshape(-1,1)
```

```python
# Create object that contains the linear model class 

model = linear_model.LinearRegression()

# Assign independent (FGA) vs dependent variable (PTS) : (PTS = intercept + coeff.FGA + error)

model.fit (x,y)

# Get the R-squared (R2) value and round it to two decimals. Variance explanation. 

r2 = round(model.score(x,y), 2)
predicted_y = model.predict(x)
```

```python
# Time for Visualisation - matplot lib

plt.scatter(x,y, s = 15, alpha = .5) # Scatterplot: Specify Size(s) and transaprency (alpha) of data
plt.plot(x, predicted_y, color = "red") # Adding a line for regression w predicted values
plt.title('NBA 2022 - FGA/PPG') # Title  
plt.xlabel('FGA per Game') # Indepdendent variable
plt.ylabel ('Points per Game') # Dependent Variable

plt.text(10,28, f'R2={r2}')  # 10,28 are the coordinates for our text to illustrate the R-squared value

ax = max(y)
xpos = np.where(y == ymax)
xmax = x[xpos]

plt.annotate(df.PLAYER[0], xy=(xmax, ymax), xytext=(xmax, ymax + 5), arrowprops=dict(arrowstyle='-'),)

plt.show()

# Seems like there are 4 "outliers" on offence - PPG. Those players have taken more than 17+ shots per game and perform better than what the model predicts.

# Most players tend to take less than 10 shots per game. 
```
